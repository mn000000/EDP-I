# Dirichlet b)

La solución de la ecuación 
$$\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = 0 $$
sujeta a las condiciones de frontera:
$$u(0,y)=10y, ~~ u_x(1,y)=-1, \;\;\; 0 < y < 1$$
$$u(x,0)=u(x,1)=0\;\;\; 0 < x < 1 $$
Esta dada en términos de la serie de Fourier por la expresión:
$$ u(x,y) = \sum_{n=1}^{\infty} \sin(n \pi y) \left[ \left( \frac{20 (-1)^{n+1}}{n \pi} \right) \cosh (n \pi x) + \left( \frac{2[(-1)^n-1]\cdot sech(n\pi)+20(-1)^n(n \pi)\tanh (n\pi)}{(n \pi)^2 } \right)\cdot  \sinh(n \pi x) \right] $$


```python
from pylab import *
from matplotlib import animation,rc 
from IPython.display import HTML 
import matplotlib.pyplot as plt
```

```python
def u_x_y(x,y,N=13):
    f = zeros( (len(x),len(y)) )
    for i in range(len(x)):
        for n in range(1,N):
            a = -20*(-1)**n/(n*pi)
            b = (20*(-1)**n*n*pi*sinh(n*pi)-2*(1-(-1)**n))/(n**2*pi**2*cosh(n*pi))
            f[:,i] += sin(n*pi*y)*(a*cosh(n*pi*x[i])+b*sinh(n*pi*x[i]))
    return f
```

Esta es una gráfica en 3D de la solución.

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection="3d")

x = linspace(0,1.0,100)
y = linspace(0,1.0,100)

X,Y = np.meshgrid(x,y)

U = u_x_y(x,y)

plt.xlabel('y'); plt.ylabel('x'); 
plt.title("u(x,y)")
ax.plot_surface(X, Y, U, cmap="viridis");
     
```
