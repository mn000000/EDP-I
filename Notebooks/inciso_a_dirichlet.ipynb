# Dirichlet a)

La solución de la ecuación 
$$\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = 0 $$
sujeta a las condiciones de frontera:
$$u(0,y)=u(1,y)=0, \;\;\; 0 < y < 1$$
$$u(x,0)=100, ~~ u(x,1)=200\;\;\;0<x < 1 $$
Esta dada en términos de la serie de Fourier por la expresión:
$$ u(x,y) = \frac{200}{\pi} \sum_{n=1}^{\infty} sen(n \pi x)⋅\frac{1+(-1)^{n+1}}{n} \left[  cosh(n\pi y)+\frac{senh(n\pi y)}{senh(n\pi)}(2-cosh(n\pi))  \right] $$


```python
from pylab import *
from matplotlib import animation,rc 
from IPython.display import HTML 
import matplotlib.pyplot as plt
```

```python
def u_x_y(x,y,N=30):
    f = zeros( (len(x),len(y)) )
    for i in range(len(y)):
        for n in range(1,N):
            a = 200*(1-(-1)**n)/(n*pi)
            b = a*(2-cosh(n*pi))/sinh(n*pi)
            f[:,i] += sin(n*pi*x)*(a*cosh(n*pi*y[i])+b*sinh(n*pi*y[i]))
    return f
```

Esta es una gráfica en 3D de la solución.

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection="3d")

x = linspace(0,1.0,100)
y = linspace(0,1.0,100)

X,Y = np.meshgrid(x,y)

U = u_x_y(x,y)

plt.xlabel('y'); plt.ylabel('x'); 
plt.title("u(x,y)")
ax.plot_surface(X, Y, U, cmap="inferno");
```
